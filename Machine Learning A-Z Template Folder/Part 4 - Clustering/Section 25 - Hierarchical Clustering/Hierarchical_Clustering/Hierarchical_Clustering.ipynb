```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import data

dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,[3,4]].values # select all rows and columns 3 and 4
```

```python
#use dendrogram to find optimal number of clusters

import scipy.cluster.hierarchy as sch
```

```python
# build dendrogram
# 'ward' method minimizes within cluster variance

# optimal clusters =5 

dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show
```

```python
# fit HC algorithm to data using 5 clusters

from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
Y_hc=hc.fit_predict(X)
```

```python
# vis results - use code from k-means

# vis the clusters - scatter plot with centroids - only for 2-dim data
# don't use for multi dim data

# second 0 indicates first column in dataset X

plt.scatter(X[Y_hc==0,0],X[Y_hc==0,1],s=100, c='red',label='Cluster 1')
plt.scatter(X[Y_hc==1,0],X[Y_hc==1,1],s=100, c='blue',label='Cluster 2')
plt.scatter(X[Y_hc==2,0],X[Y_hc==2,1],s=100, c='green',label='Cluster 3')
plt.scatter(X[Y_hc==3,0],X[Y_hc==3,1],s=100, c='cyan',label='Cluster 4')
plt.scatter(X[Y_hc==4,0],X[Y_hc==4,1],s=100, c='magenta',label='Cluster 5')

# don't plot centroids since agglomerative doesn't use centroids

plt.title('Clusters of clients')
plt.xlabel('Anuual Income')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show
```
