```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,2:].values
```

```python
# very little data so won't split into training and testing sets

'''from sklearn.linear_model import 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)'''
```

```python
# don't need feature scaling - using same library as LR which scales automatic
```

```python
# fit linear regression to dataset - for comparison

from sklearn.linear_model import LinearRegression
Lin_reg=LinearRegression()
Lin_reg.fit(X,Y)
```

```python
#build polynomial regression - import new class and transform X into new matrix
# to include X and X squared

from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=2)
X_poly=poly_reg.fit_transform(X)
```

```python
# fit new polynomial

Lin_reg2=LinearRegression()
Lin_reg2.fit(X_poly,Y)
```

```python
# visualize the linear regression results
plt.scatter(X,Y,color='red') #real data
plt.plot(X,Lin_reg.predict(X),color='blue') # predicted solaries as a line
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,Lin_reg2.predict(poly_reg.fit_transform(X))) 

#Xpoly already defined
# for existing features of X. Want to make the code general to any matrix
# of X

plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show
```

```python
# change the degree of the polynomial

from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=3)
X_poly=poly_reg.fit_transform(X)

Lin_reg2=LinearRegression()
Lin_reg2.fit(X_poly,Y)

plt.scatter(X,Y,color='red')
plt.plot(X,Lin_reg2.predict(poly_reg.fit_transform(X))) 

#Xpoly already defined
# for existing features of X. Want to make the code general to any matrix
# of X

plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show
```

```python
# change the degree of the polynomial

from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)

Lin_reg2=LinearRegression()
Lin_reg2.fit(X_poly,Y)

plt.scatter(X,Y,color='red')
plt.plot(X,Lin_reg2.predict(poly_reg.fit_transform(X))) 

#Xpoly already defined
# for existing features of X. Want to make the code general to any matrix
# of X

plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show
```
